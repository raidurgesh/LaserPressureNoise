```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 30000)
RIN_PSD_o=(1e-4)**2*(1+(0.1e-3/freq)**2)
RIN_PSD_n=(1e-4)**2*(1+(0.5e-3/freq))
plt.style.use('bmh')
plt.plot(freq, RIN_PSD_o, 'r-',label = "Old")
plt.plot(freq, RIN_PSD_n, 'b-',label = "New")
plt.yscale("log")
plt.xscale("log")
plt.title("RIN in the LISA Band PSD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("RIN (1/(Hz)")
plt.legend(title = "PSD")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 30000)
RIN_ASD_o=(1e-4)*(1+(0.1e-3/freq)**2)**0.5
RIN_ASD_n=(1e-4)*(1+(0.5e-3/freq))**0.5
plt.style.use('bmh')
plt.plot(freq, RIN_ASD_o, 'r-',label = "Old")
plt.plot(freq, RIN_ASD_n, 'b-',label = "New")
plt.yscale("log")
plt.xscale("log")
plt.title("RIN in the LISA Band ASD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("RIN (1/(sqrtHz)")
plt.legend(title = "ASD")
plt.show()
```
