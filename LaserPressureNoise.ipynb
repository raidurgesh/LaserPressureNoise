```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 30000)
RIN_PSD=(1e-4)**2*(1+(0.1e-3/freq)**2)
plt.style.use('bmh')
plt.plot(freq, RIN_PSD, 'r-',label = "RIN")
plt.yscale("log")
plt.xscale("log")
plt.title("RIN in the LISA Band PSD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("RIN (1/(Hz)")
plt.legend(title = "PSD")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 30000)
RIN_ASD=(1e-4)*(1+(0.1e-3/freq))
plt.style.use('bmh')
plt.plot(freq, RIN_ASD, 'b-',label = "ASD")
plt.yscale("log")
plt.xscale("log")
plt.title("RIN in the LISA Band ASD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("RIN (1/sqrt(Hz)")
plt.legend(title = "ASD")
plt.show()
```
